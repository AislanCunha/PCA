# Principal Component Analysis - PCA

### Import the libraries

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

### Import the data

```python
df = pd.read_csv('data_PCA_std.csv')
print (df)
```

### Correlation Matriz

```python
df_corr = df.corr()
df_corr
```

```python
sns.heatmap(df_corr, xticklabels=df_corr.columns, yticklabels=df_corr.columns, cmap="YlGnBu")
```

### Decomposition of the variance and covariance matrix into principal components

```python
X = np.matrix(df)
S = np.cov(np.transpose(X))
```

```python
pca = PCA(n_components=5)
```

```python
pca
```

```python
pca.fit(X)
```

```python
pca.components_[0,:]
```

```python
pca.explained_variance_ratio_
np.round(pca.explained_variance_ratio_,3)
```

```python
fig = plt.figure(figsize=(8,5))

plt.plot(pca.explained_variance_ratio_, 'ro-', linewidth=2)
plt.show()
```

```python
pca.transform(X)
```

```python
PCA1 = pca.transform(X)[:,0]
```

```python
df['PCA1'] = PCA1
```
